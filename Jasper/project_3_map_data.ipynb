```python
import os
import csv
import json
import pandas as pd
import geojson
```

```python
data = "Resources/food_data/business_final.csv"
```

```python
data_df = pd.read_csv(data, encoding="ISO-8859-1")
```

```python
data_df.head()
```

```python
data_dropped_df = data_df.drop(['business_id', 'is_open', 'attr_key', 'attr_value'], axis = 1)
data_dropped_df
```

```python
data_dropped_dup_df = data_dropped_df.drop_duplicates(subset = ['name'])
data_dropped_dup_df.to_csv(r'business_json.csv', index = False)
data_dropped_dup_df
```

```python
new_data_df = data_dropped_dup_df[data_dropped_dup_df.state != "BC"]
new_data_df
```

```python
new_data_df.to_csv('business_edit.csv', encoding='utf-8')
```

```python
result = new_data_df.to_json(orient="records")
parsed = json.loads(result)
business_json = json.dumps(parsed, indent=4)
jsonFile = open("business.json", "w")
jsonFile.write(business_json)
jsonFile.close()
```
